```python


def to_csv(fname,max):
    train_image = open('train-images.idx3-ubyte','rb')
    train_label = open('train-labels.idx1-ubyte','rb')

   
    csv_file = open(fname+".csv",'w') 

    mag, img_length = struct.unpack(">II",train_image.read(8)) 
    mag, label_length = struct.unpack(">II",train_label.read(8)) 

    rows, cols = struct.unpack(">II",train_image.read(8)) 
    pixels = rows * cols 
    for idx in range(img_length): 
        if idx > max : break 
       
        label = struct.unpack("B",train_label.read(1))[0] 
   
        bdata = train_image.read(pixels) 
    
        sdata = list (map(lambda a :str(a),bdata)) 
        csv_file.write(str(label)+',') 
        csv_file.write(','.join(sdata)+"\n") 
    train_label.close() 
    train_image.close() 
    csv_file.close() 


to_csv('train',10000)

```

```python
def load_csv(fname):
    labels = []
    images = []
    with open(fname , 'r') as f:
        for line in f:

            cols = line.split(',')
            if len(cols) < 2 : continue

            labels.append(int(cols.pop(0)))
            vals = list(map(lambda a:int(a)/256,cols))
            images.append(vals)
    return {"labels":labels,"images":images}



data = load_csv("./train.csv")


```

```python
def make_u1_u2(d,p,c):
    u1=np.random.uniform(-1.0, 1.0, size=(p+1)*(d+1))
    u1=u1.reshape((p+1),(d+1))
    u2=np.random.uniform(-1.0, 1.0, size=(c+1)*(p+1))
    u2=u2.reshape((c+1),(p+1))
    print("가중치 행렬 초기화 : u1:",u1.shape,", u2:",u2.shape)
    return u1,u2
d=784
p=50
c=3
u1,u2=make_u1_u2(d,p,c)
uu=u1
uu.shape
```

```python
def one_hot(y):
    a=np.zeros((len(y),3))
    for i in range(len(y)):
        if y[i]==1:
            a[i][0]=1
        elif y[i] ==5:
            a[i][1]=1
        else:
            a[i][2]=1
    return a
    
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def derivative_of_sigmoid(x):
    return sigmoid(x)*(1-sigmoid(x))
```

```python
def algo1(x_train,y_train):
    d=784
    p=50
    c=3
    epoch=2
    lr=0.001 #learning rate
  
    u1,u2=make_u1_u2(d,p,c) #return u1,u2
    
    
    while(epoch>0):
        epoch=epoch-1
        print("epoch:",epoch)
        pm = np.random.permutation(x_train.shape[0])
        x_train = x_train[pm] # shuffle x

        for idx in range(x_train.shape[0]): 
            x=x_train[idx]
            x=np.append([1],x)  # x0=1
            z=np.ones(p+1) # z0=1


            o=np.ones(c+1) # output layer
            y=y_train[idx] #Result
            y=np.append([1],y)
            zsum=np.zeros(p+1)

            osum=np.zeros(c+1) 
            n=np.zeros(p+1) 

            sig=np.zeros(c+1) 

            triang1e1=u1
            triang1e2=u2
            
            #Forward Calculation

            for j in range(1,p+1):
                
                zsum[j]=np.dot(u1[j],x)
                z[j]=(sigmoid(zsum[j]))
                
            for k in range(1,c+1):
                
                osum[k]=np.dot(u2[k],z)
                o[k]=sigmoid(osum[k])
                
            #error reverse propagation
            for k in range(1,c+1):
                sig[k]=(y[k]-o[k])*derivative_of_sigmoid(osum[k])
            for k in range(1,c+1):
                for j in range(0,p+1):
                    triang1e2[k][j]=-sig[k]*z[j]
            for j in range(1,p+1): 
                n[j]=derivative_of_sigmoid(zsum[j])
                sum=0
                for q in range(1,c+1):
                    sum=sum+(sig[q]*u2[q][j])
                n[j]=n[j]*sum
            for j in range(1,p+1):
                for i in range(0,d+1):
                    triang1e1[j][i]=-n[j]*x[i]
                    
            #weight renewal
            for k in range(1,c+1):
                for j in range(0,p+1):
                    u2[k][j]=u2[k][j]-lr*triang1e2[k][j]
            for j in range(1,p+1):
                for i in range(0,d+1):
                    u1[j][i]=u1[j][i]-lr*triang1e1[j][i]
              
        print(u1)
        print(u2)
    return (u1,u2)

```
